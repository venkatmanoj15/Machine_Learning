```python
from sklearn.preprocessing import LabelEncoder

workclass_encoder = LabelEncoder()
df1['workclass'] = workclass_encoder.fit_transform(df1['workclass'])

education_encoder = LabelEncoder()
df1['education'] = education_encoder.fit_transform(df1['education'])

marital_encoder = LabelEncoder()
df1['maritalstatus'] =marital_encoder.fit_transform(df1['maritalstatus'])

occupation_encoder = LabelEncoder()
df1['occupation'] =occupation_encoder.fit_transform(df1['occupation'])

relationship_encoder = LabelEncoder()
df1['relationship'] =relationship_encoder.fit_transform(df1['relationship'])

race_encoder = LabelEncoder()
df1['race'] =race_encoder.fit_transform(df1['race'])

sex_encoder = LabelEncoder()
df1['sex'] =sex_encoder.fit_transform(df1['sex'])

native_encoder = LabelEncoder()
df1['native'] = native_encoder.fit_transform(df1['native'])

salary_encoder = LabelEncoder()
df1['Salary'] = salary_encoder.fit_transform(df1['Salary'])

```

```python
x=df.drop('Salary',axis=1)
y=df['Salary']
```

```python
x_test=df1.drop('Salary',axis=1)
y_test=df1['Salary']
```

```python
def logistic_regression_model():
    from sklearn.linear_model import LogisticRegressionCV
    model = LogisticRegressionCV()
    model.fit(x,y)
    return model

def naive_bayes_model():
    from sklearn.naive_bayes import GaussianNB
    model = GaussianNB()
    model.fit(x,y)
    return model

def knn_model():
    from sklearn.neighbors import KNeighborsClassifier
    model = KNeighborsClassifier()
    model.fit(x,y)
    return model

def svm_model():
    from sklearn.svm import SVC
    model = SVC()
    model.fit(x,y)
    return model

def decision_tree_model():
    from sklearn.tree import DecisionTreeClassifier
    model = DecisionTreeClassifier()
    model.fit(x,y)
    return model

def random_forest_model():
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier()
    model.fit(x,y)
    return model

def adaboost_model():
    from sklearn.ensemble import AdaBoostClassifier
    model = AdaBoostClassifier()
    model.fit(x,y)
    return model
    
```

```python
models = [
    {"model_name":"Logistic Regression","model":logistic_regression_model()},
    {"model_name":"Naive Bayes","model":naive_bayes_model()},
    {"model_name":"KNN","model":knn_model()},
    {"model_name":"SVM","model":svm_model()},
    {"model_name":"Decision Tree","model":decision_tree_model()},
    {"model_name":"Random Forest","model":random_forest_model()},
    {"model_name":"ADA Boost","model":adaboost_model()}
]
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
def evaluate_model(model,model_name,x,y):
    y_pred = model.predict(x)
    
    accuracy = accuracy_score(y,y_pred)
    precision = precision_score(y,y_pred)
    recall = recall_score(y,y_pred)
    f1 = f1_score(y,y_pred)
    
    cm = confusion_matrix(y,y_pred)
    tn,fp,fn,tp = cm.ravel()
    accuracy_cm = (tn+tp)/(tn+tp+fn+fp)

      # return the statistics
    return model_name, accuracy, precision, recall, f1, accuracy_cm
    
```

```python
# compare the models by evaluating them
metrics_train = []
for model_info in models:
    # evaluate the model and collect the metrics
    metrics_train.append(evaluate_model(model_info['model'], model_info['model_name'], x, y))
# convert the list of tuples to a data frame
pd.DataFrame(metrics_train, columns=['Algorithm', 'Accuracy', 'Precision', 'Recall', 'F1 Score','CM Accuracy'])
```

```python
# compare the models by evaluating them
metrics_test = []
for model_info in models:
    # evaluate the model and collect the metrics
    metrics_test.append(evaluate_model(model_info['model'], model_info['model_name'], x_test, y_test))
# convert the list of tuples to a data frame
pd.DataFrame(metrics_test, columns=['Algorithm', 'Accuracy', 'Precision', 'Recall', 'F1 Score','CM Accuracy'])
```
