```python
# Que1.-
# In this dataset you have to predict that whether we have to give ATTORNEY to the clients or not (0=Yes, 1=No) based on below factors.
# CLMAGE - Age of Client
# CLMSEX - Sex of client (0=Male, 1= Female)
# SEATBELT - Whether client has Seatbelt or not (0=Yes, 1=No)
# CLMINSUR - Whether Client give Insurance or not (0=Yes, 1=No)
# LOSS - Loss that client faced

# Prepare model by using Naive Bayes,KNN and SVM & Also calculate Accuracy by confusion matrix, Calculate accuracy score, precision score, recall score, f1 score.

# Dataset : Claimants.csv
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('claimants.csv')
df.head(2)
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
x=df.drop('ATTORNEY',axis=1)
y=df['ATTORNEY']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
def create_logistic_regression_model():
    from sklearn.linear_model import LogisticRegressionCV
    model = LogisticRegressionCV()
    model.fit(x_train,y_train)
    return model
```

```python
def create_naive_bayes_model():
    from sklearn.naive_bayes import GaussianNB
    model = GaussianNB()
    model.fit(x_train,y_train)
    return model
```

```python
def create_knn_model():
    from sklearn.neighbors import KNeighborsClassifier
    model = KNeighborsClassifier(n_neighbors=3)
    model.fit(x_train,y_train)
    return model
```

```python
def create_svm_model():
    from sklearn.svm import SVC
    model = SVC()
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

def evaluate_model(model,model_name):
    y_pred = model.predict(x_test)

    accuracy = accuracy_score(y_test,y_pred)
    precision = precision_score(y_test,y_pred)
    recall = recall_score(y_test,y_pred)
    f1 = f1_score(y_test,y_pred)
    
    return model_name,accuracy,precision,recall,f1
    
```

```python
from sklearn.metrics import confusion_matrix

def confusion_matrix1(model,model_name):
    y_pred = model.predict(x_test)
    cm=confusion_matrix(y_test,y_pred) 
    tn, fp, fn, tp = cm.ravel() 

    return {
        'model_name': model_name,
        'matrix': cm,
        'true_negatives': tn,
        'false_positives': fp,
        'false_negatives': fn,
        'true_positives': tp
    }
```

```python
models = [
    {"model_name":"Logistic Regression","model":create_logistic_regression_model()},
    {"model_name":"Naive Bayes","model":create_naive_bayes_model()},
    {"model_name":"KNN","model":create_knn_model()},
    {"model_name":"SVM","model":create_svm_model()}
]
```

```python
metrics=[]

for model_info in models:
    stats = evaluate_model(model_info['model'],model_info['model_name'])
    metrics.append(stats)
```

```python
pd.DataFrame(metrics,columns=['Algorithm','Accuracy','Precision','Recall','F1 Score'])
```

```python
##confusion_matrix
metrics1=[]

for model_info in models:
    stats = confusion_matrix1(model_info['model'],model_info['model_name'])
    stats
    metrics1.append(stats)
```

```python
df2 =pd.DataFrame(metrics1)
df2
```

```python
#accuracy by confusion matrix
metrics2=[]

for model_info in models:
    stats = confusion_matrix1(model_info['model'],model_info['model_name'])
    stat={'model_name':stats['model_name'],'accuracy_by_cm':(stats['true_positives']+stats['true_negatives'])/(stats['true_positives']+stats['true_negatives']+stats['false_negatives']+stats['false_positives'])}
    print(stat)
    metrics2.append(stat)
    
```

```python
df3 =pd.DataFrame(metrics2)
df3
```
